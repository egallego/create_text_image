```python
from generator import ImageCreator
from os.path import join
from IPython.display import display
```

## Parameters

```python
path_results='./results'
```

## Code
### Example image 1
Create an example image with the "Hellow World" text

```python
# Initialize object
image_creator=ImageCreator(font='font_example.ttf', xmin=90, font_size=80)

# Define text to add to the image
image_creator.create_image("Hello World!")

# Display image
display(image_creator.image)

# Store Image
image_creator.save_image(join(path_results, 'image_example_1.jpg'))
```

### Example image 2
Create a second example image with a longer text and an Author text

```python
# Initialize object
image_creator=ImageCreator(name='Author', font='font_example.ttf')

# Define text to add to the image
image_creator.create_image("Example of how to create an image")

# Display image
display(image_creator.image)

# Store Image
image_creator.save_image(join(path_results, 'image_example_2.jpg'))
```

### Example image 3
Create an example image with the white background


```python
# Initialize object
image_creator=ImageCreator(name='Author', font='font_example.ttf', color='white')

# Define text to add to the image
image_creator.create_image("Example of how to create an image with white background")

# Display image
display(image_creator.image)

# Store Image
image_creator.save_image(join(path_results, 'image_example_3.jpg'))
```
